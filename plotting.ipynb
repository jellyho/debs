```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bootstrap
# -----------------------------------------------------------
# 1. CI 계산 함수
# -----------------------------------------------------------
def calculate_scipy_ci(data, ci=95):
    data = np.array(data)
    data = data[~np.isnan(data)]
    
    if len(data) < 2 or np.all(data == data[0]):
        mean_val = np.mean(data)
        return mean_val, mean_val

    res = bootstrap(
        (data,), 
        statistic=np.mean, 
        confidence_level=ci/100, 
        n_resamples=1000, 
        method='percentile'
    )
    return res.confidence_interval.low, res.confidence_interval.high
target_step = 1000000
alpha_df_new = pd.read_csv("plots/alpha_tuning_new.csv")
filtered_df = alpha_df_new[alpha_df_new['_step'].astype(int) == target_step].copy()
filtered_df = filtered_df[filtered_df['config.agent.alpha'] > 0.01]
filtered_df = filtered_df[filtered_df['config.task_name'].isin(['puzzle-3x3-play', 'scene-play', 'cube-double-play'])]
filtered_df = filtered_df[
    (filtered_df['group_id'].isin(['dsrl', 'mflql']) & filtered_df['config.agent.latent_dist'].isin(['sphere'])) | 
    (filtered_df['group_id'].isin(['qcfql']) & filtered_df['config.agent.latent_dist'].isin(['normal']))
]

name_map = {
    'mflql': 'LPS',    # 2
    'dsrl': 'DSRL', # 1
    'qcfql': 'QC-MFQL', # 0  
    # 'qcfql': 'QC-FQL'
}

color_map = {
    'LPS' : 0,    # 2
    'DSRL': 1, # 1
    'QC-MFQL': 8, # 0
    'QC-FQL': 7, # 0  
}
filtered_df['group_id'] = filtered_df['group_id'].replace(name_map)
tasks = filtered_df['config.task_name'].unique()

desired_order = ['QC-MFQL', 'DSRL', 'LPS']

existing_groups = set(filtered_df['group_id'].unique())
groups = [g for g in desired_order if g in existing_groups]

colors = sns.color_palette("husl", 12)
markers = ['o', 's', 'D', '^', 'v', 'p'] # 원, 사각형, 다이아몬드, 삼각형...

# 그룹별 스타일 매핑 (모든 플롯에서 일관성 유지)
style_map = {g: {'color': colors[color_map[g]], 'marker': markers[color_map[g] % len(markers)]} for i, g in enumerate(groups)}

# -----------------------------------------------------------
# 3. Plotting (Subplots)
# -----------------------------------------------------------
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True) 

legend_handles = []
legend_labels = []

for i, task in enumerate(tasks):
    ax = axes[i]
    task_df = filtered_df[filtered_df['config.task_name'] == task]
    
    for group_name in groups:
        group_df = task_df[task_df['group_id'] == group_name]
        if group_df.empty: continue
        
        agg_df = group_df.groupby('config.agent.alpha')['eval/success'].apply(list).reset_index(name='values')
        agg_df = agg_df.sort_values('config.agent.alpha')

        x_list = []
        means = []
        lowers = []
        uppers = []

        for _, row in agg_df.iterrows():
            vals = row['values']
            mean_val = np.mean(vals)
            lower, upper = calculate_scipy_ci(vals, ci=95)
            
            x_list.append(row['config.agent.alpha'])
            means.append(mean_val)
            lowers.append(lower)
            uppers.append(upper)

        x = np.array(x_list)
        mean = np.array(means)
        lower = np.array(lowers)
        upper = np.array(uppers)

        style = style_map[group_name]

        line, = ax.plot(x, mean, 
                        label=group_name, 
                        color=style['color'], 
                        marker=style['marker'], 
                        markersize=16,           
                        linewidth=5)
        
        ax.fill_between(x, lower, upper, color=style['color'], alpha=0.15)
        
        if i == 0:
            legend_handles.append(line)
            legend_labels.append(group_name)

    # --- [수정된 부분] ---
    
    # 1. Title을 \texttt{} 스타일(monospace)로 변경
    if task == "cube-double-play":
        title = "cube-double"
    elif task == "scene-play":
        title = "scene"
    elif task == "puzzle-3x3-play":
        title = "puzzle-3x3"
    # title = f"{task}{'-sparse' if task in ['scene-play', 'puzzle-3x3-play'] else ''}-task{2 if task in ['cube-double-play', 'scene-play'] else 4}"
    ax.set_title(title, fontsize=30, fontfamily='monospace')
    
    ax.set_xlabel("Alpha (α)", fontsize=20)
    ax.set_xscale('log') 
    
    # 2. Grid를 'major'로 변경하여 보조선 개수 줄임
    ax.grid(True, which="major", ls="-", alpha=0.5)
    
    if i == 0:
        ax.set_ylabel("Success Rate", fontsize=20)

    ax.tick_params(axis='both', which='major', labelsize=20)

# -----------------------------------------------------------
# 4. 마무리
# -----------------------------------------------------------
plt.tight_layout()

fig.legend(legend_handles, legend_labels, 
            loc='lower center', 
            bbox_to_anchor=(0.5, -0.12), 
            ncol=len(groups),           
            fontsize=25, 
            frameon=True)
plt.savefig(f'plots/alpha_plot.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import re

def generate_latex_rows(df, visual_df=None):
    # 1. 데이터 전처리 (Step 1M 필터링)
    df_filtered = df[df['_step'] == 1000000].copy()
    df_filtered['eval/success'] = df_filtered['eval/success'] * 100
    df_filtered['model_key'] = df_filtered['group_id'].str.split('_').str[1]
    
    if visual_df is not None:
        visual_filtered = visual_df[visual_df['_step'] == 500000].copy()
        visual_filtered['eval/success'] = visual_filtered['eval/success'] * 100
        # visual_df의 group_id 형식이 다를 경우를 위해 예외처리
        visual_filtered['model_key'] = visual_filtered['group_id']
        # .loc[행_조건, 열_이름] = 바꿀_값
        visual_filtered.loc[visual_filtered['model_key'] == 'mflql', 'model_key'] = 'meanflowq'

    # 모델 매핑 및 순서
    col_mapping = {
        'flow': 'BC-FM', 'meanflow': 'BC-MF',
        'qcfql': 'QC-FQL', 'qcmfql': 'QC-MFQL',
        'fmlql': 'FM-LQL', 'dsrl': 'DSRL',
        'meanflowq': 'LPS', 'distill': 'MF-LDQL', 'cfgrl': 'CFGRL'
    }
    target_columns = ['BC-FM', 'BC-MF', 'QC-FQL', 'QC-MFQL', 'CFGRL', 'DSRL', 'LPS']

    # Task 정의 (Visual 추가)
    tasks = [
        "cube-single-play", "cube-double-play", "scene-play",
        "puzzle-3x3-play", "puzzle-4x4-play", "visual-*-task1"
    ]

    def get_stats(data_subset):
        n = len(data_subset)
        if n == 0: return None, "-"
        m = data_subset['eval/success'].mean()
        s = (data_subset['eval/success'].std() / np.sqrt(n)) if n > 1 else 0
        return m, f"${m:.0f} \\pm {s:.0f}$"

    latex_rows = []

    for task in tasks:
        display_name = f"\\texttt{{{task}}}" if task == "visual-*-task1" else \
                       f"\\texttt{{{task}{'-sparse' if task in ['scene-play', 'puzzle-3x3-play'] else ''}-singletask}}"
        
        row_data = [] # (mean_val, formatted_string) 튜플 저장
        
        for col_header in target_columns:
            curr_model_keys = [k for k, v in col_mapping.items() if v == col_header]
            if not curr_model_keys:
                row_data.append((None, "-"))
                continue
            
            curr_model_key = curr_model_keys[0]
            
            # --- Task 별 데이터 필터링 ---
            if task == "visual-*-task1" and visual_df is not None:
                task_df = visual_filtered[visual_filtered['model_key'] == curr_model_key]
                # print(task_df['group_id'].unique())
            else:
                model_df = df_filtered[df_filtered['model_key'] == curr_model_key]
                base_task = '-'.join(task.split('-')[:3])
                task_df = model_df[model_df['config.task_name'] == base_task]

            # Normal / Sphere 각각 계산
            m_norm, s_norm = get_stats(task_df[task_df['config.agent.latent_dist'] == 'normal'])
            m_sphere, s_sphere = get_stats(task_df[task_df['config.agent.latent_dist'] == 'sphere'])

            # 비교를 위한 최대값 추출 (Normal과 Sphere 중 더 높은 것)
            max_m = max([v for v in [m_norm, m_sphere] if v is not None], default=-1)
            
            if s_norm == "-" and s_sphere == "-":
                row_data.append((None, "-"))
            else:
                row_data.append((max_m, f"{s_norm} / {s_sphere}"))

        # --- 최고 성능 강조 (Bold) 로직 ---
        # 이번 행(Task)에서 가장 높은 평균값 찾기
        valid_means = [d[0] for d in row_data if d[0] is not None]
        best_val = max(valid_means) if valid_means else -1
        
        formatted_cells = []
        for mean_val, cell_str in row_data:
            if mean_val is not None and mean_val == best_val and best_val > 0:
                # $숫자 \pm 숫자$ 패턴을 찾아 \textbf{숫자 \pm 숫자}로 변경
                # 슬래시(/) 전후의 모든 수식에 볼드를 입힙니다.
                bold_cell = cell_str.replace("$", "$\\mathbf{")
                bold_cell = bold_cell.replace("}$", "}$") # LaTeX 수식 내부 볼드 처리
                # 더 확실한 방법: 수식 기호 전체를 볼드체로 감싸기
                cell_str = f"\\textbf{{{cell_str}}}"
            formatted_cells.append(cell_str)

        latex_rows.append(f"{display_name} & " + " & ".join(formatted_cells) + " \\\\")

    return "\n".join(latex_rows)

# 실행 예시
df = pd.read_csv("plots/OGBench_results.csv")
v_df = pd.read_csv("plots/visual.csv")
print(generate_latex_rows(df, v_df))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bootstrap
import matplotlib.patches as mpatches
import matplotlib.colors as mc
import colorsys

# -----------------------------------------------------------
# [Helper] Style & CI Functions
# -----------------------------------------------------------
def calculate_scipy_ci(data, ci=95):
    data = np.array(data)
    data = data[~np.isnan(data)]
    if len(data) < 2 or np.all(data == data[0]):
        mean_val = np.mean(data)
        return mean_val, mean_val
    res = bootstrap((data,), statistic=np.mean, confidence_level=ci/100, 
                    n_resamples=1000, method='percentile')
    return res.confidence_interval.low, res.confidence_interval.high

def adjust_lightness(color, amount=0.5):
    try: c = mc.cnames[color]
    except: c = color
    c = mc.to_rgb(c)
    c = colorsys.rgb_to_hls(*c)
    return colorsys.hls_to_rgb(c[0], max(0, min(1, amount * c[1])), c[2])

# -----------------------------------------------------------
# 1. Data Loading
# -----------------------------------------------------------
try:
    df = pd.read_csv("plots/OGBench_results.csv")
    df['group_id'] = df['group_id'].str.split('_').str[1]
except:
    print("Warning: Main CSV not found. Creating dummy data.")
    df = pd.DataFrame({
        'group_id': ['fmlql']*10 + ['distill']*10 + ['mflql']*10 + ['meanflowq']*10,
        'config.agent.latent_dist': ['sphere']*40,
        '_step': [100]*40,
        'eval/success': np.random.rand(40)
    })

try:
    orig_df = pd.read_csv('plots/orig.csv')
    onestep_df = pd.read_csv('plots/lqlonestep.csv')
    df = pd.concat([df, orig_df, onestep_df], ignore_index=True)
except FileNotFoundError:
    print("Warning: Additional CSVs not found. Using main/dummy data.")

# df.loc[조건, 바꿀컬럼명] = 변경할값
df.loc[df['group_id'] == 'original', 'eval/success'] = 0.0

# 이름 매핑
name_map = {
    'fmlql': 'FM-LPS',
    'fmonesteplql': 'FM-1step-LPS',
    'original': 'LPS (Original MF)',
    'meanflowq': 'LPS'
}
df['group_id'] = df['group_id'].replace(name_map)


# [색상 및 패턴 정의]
colors = sns.color_palette("husl", 12)
main_color = colors[0]  # LQL (Ours) - 빨강 계열
variant_color = 'lightgray' # 나머지 변형 - 회색

method_colors = {
    'FM-LPS': variant_color,
    'FM-1step-LPS': variant_color,
    'LPS (Original MF)': variant_color,
    'LPS': main_color
}

method_hatches = {
    'FM-LPS': '///',        # 빗금
    'FM-1step-LPS': '...',  # 점박이
    'LPS (Original MF)': 'xxx', # 격자
    'LPS': ''               # 민무늬 (강조)
}

# -----------------------------------------------------------
# 2. Plotting Setup
# -----------------------------------------------------------
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    'font.family': 'monospace', 
    'font.size': 18,       
    'axes.labelsize': 20, 
    'xtick.labelsize': 18, 
    'ytick.labelsize': 18, 
    'figure.titlesize': 24,
    'legend.fontsize': 16
})

fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# -----------------------------------------------------------
# Plot 1: Effect of Base Policy (FM vs FM-1step vs LQL)
# -----------------------------------------------------------
ax1 = axes[0]
methods_1 = ['FM-LPS', 'FM-1step-LPS', 'LPS']
target_dist = 'sphere'

df_1 = df[df['group_id'].isin(methods_1) & (df['config.agent.latent_dist'] == target_dist)].copy()

records_1 = []
if not df_1.empty:
    for m in methods_1:
        subset = df_1[df_1['group_id'] == m]
        if subset.empty: continue
        agg = subset.groupby('_step')['eval/success'].apply(list).reset_index(name='values').sort_values('_step')
        if not agg.empty:
            tail_values = agg.tail(5)['values'].tolist()
            for v_list in tail_values:
                for val in v_list:
                    records_1.append({'Method': m, 'Success Rate': val})
df_plot_1 = pd.DataFrame(records_1)

if not df_plot_1.empty:
    sns.barplot(
        data=df_plot_1, x="Method", y="Success Rate",
        palette=method_colors,
        errorbar=('ci', 95), capsize=0.1, ax=ax1, width=0.6, linewidth=0
    )
    
    for i, bar in enumerate(ax1.patches):
        if i < len(methods_1):
            m_name = methods_1[i]
            bar.set_hatch(method_hatches[m_name])
            
            face_color = bar.get_facecolor()
            edge_color = adjust_lightness(face_color, amount=0.6)
            bar.set_edgecolor(edge_color)
            bar.set_linewidth(3.0)

    ax1.set_title("Effect of Base Policy", fontsize=20, fontweight='bold', pad=15)
    ax1.set_xlabel("")
    ax1.set_ylim(0, 1.05)
    # [수정] 'Variant' 제거 및 간소화
    ax1.set_xticklabels(['FM-LPS', 'FM-1step\nLPS', 'LPS\n(Ours)'])

# -----------------------------------------------------------
# Plot 2: MeanFlow Reformulation (Original vs Ours)
# -----------------------------------------------------------
ax2 = axes[1]
methods_2 = ['LPS (Original MF)', 'LPS']
df_2 = df[df['group_id'].isin(methods_2) & (df['config.agent.latent_dist'] == target_dist)].copy()

records_2 = []
if not df_2.empty:
    for m in methods_2:
        subset = df_2[df_2['group_id'] == m]
        if subset.empty: continue
        agg = subset.groupby('_step')['eval/success'].apply(list).reset_index(name='values').sort_values('_step')
        if not agg.empty:
            tail_values = agg.tail(5)['values'].tolist()
            for v_list in tail_values:
                for val in v_list:
                    records_2.append({'Method': m, 'Success Rate': val})
df_plot_2 = pd.DataFrame(records_2)

if not df_plot_2.empty:
    sns.barplot(
        data=df_plot_2, x="Method", y="Success Rate",
        palette=method_colors,
        errorbar=('ci', 95), capsize=0.1, ax=ax2, width=0.5, linewidth=0
    )
    
    for i, bar in enumerate(ax2.patches):
        if i < len(methods_2):
            m_name = methods_2[i]
            bar.set_hatch(method_hatches[m_name])
            
            face_color = bar.get_facecolor()
            edge_color = adjust_lightness(face_color, amount=0.6)
            bar.set_edgecolor(edge_color)
            bar.set_linewidth(3.0)

    ax2.set_title("MeanFlow Reformulation", fontsize=20, fontweight='bold', pad=15)
    ax2.set_xlabel("")
    ax2.set_ylabel("")
    ax2.set_ylim(0, 1.05)
    # [수정] 'Variant' 제거 및 간소화
    ax2.set_xticklabels(['LPS\n(Original MF)', 'LPS\n(Ours)'])

# -----------------------------------------------------------
# Global Styling
# -----------------------------------------------------------
plt.tight_layout()
plt.subplots_adjust(wspace=0.2)

plt.savefig('plots/ablation_base_reformulation_clean.pdf', dpi=300, bbox_inches='tight')
print("Saved plots/ablation_base_reformulation_clean.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 데이터 생성 (범위를 조금 넓게 잡음)
x = np.linspace(-4, 4, 500)
y = np.linspace(-4, 4, 500)
X, Y = np.meshgrid(x, y)

# 2. 여러 개의 모드(봉우리)를 만드는 함수 정의
# (x0, y0)를 중심으로 하는 가우시안 함수를 만들어주는 도우미 함수
def gaussian_peak(x, y, x0, y0, sigma, amplitude):
    return amplitude * np.exp(-((x - x0)**2 + (y - y0)**2) / (2 * sigma**2))

# 여러 위치에 봉우리를 배치하고 합칩니다.
# 중앙 봉우리 (넓고 낮게 퍼짐)
Z = gaussian_peak(X, Y, 0, 0, 2.0, 1.5)
# 네 구석의 봉우리 (좁고 높게)
Z += gaussian_peak(X, Y, 2., 2., 1.0, 2)
Z += gaussian_peak(X, Y, -2., 2., 1.0, 2)
Z += gaussian_peak(X, Y, -2., -2., 1.0, 2)
Z += gaussian_peak(X, Y, 2., -2., 1.0, 2)

Z += gaussian_peak(X, Y, 0, 2.5, 1.5, -1)
Z += gaussian_peak(X, Y, 0, -2.5, 1.5, -1)
Z += gaussian_peak(X, Y, 2.5, 0, 1.5, -1)
Z += gaussian_peak(X, Y, -2.5, 0, 1.5, -1)


# 3. 그림 설정
fig, ax = plt.subplots(figsize=(6, 6))
fig.patch.set_facecolor('white') # 배경 흰색
ax.set_facecolor('white')

# 4. 스타일 복원 (이전 코드의 설정 유지)
# 마음에 들어 하셨던 파란색 색상 팔레트
custom_colors = ['#bdd7ee', '#8faadc', '#5b8dd8', '#4472c4']
# 가장 연한 민트 -> 짙은 청록색
# custom_colors = # 흰색에 가까운 하늘색 -> 부드러운 파스텔 블루
custom_colors = soft_baby_blue = ['#f5faff', '#e3f2fd', '#bbdefb', '#90caf9']

# 연한 라벤더 핑크 -> 딥 퍼플
# custom_colors = pink_purple = ['#f3e5f5', '#ce93d8', '#8e24aa', '#4a148c']

# 아이스 블루 -> 클래식 블루 -> 미드나잇 블루
# custom_colors = deep_ocean = ['#e3f2fd', '#64b5f6', '#1976d2', '#0d47a1']
# 새로운 함수 Z의 값 범위에 맞춰 등고선 레벨 재설정
# 이 값들을 조절하면 봉우리의 크기와 연결 정도가 달라집니다.
levels = [0.5, 0.9, 1.3, 1.7]

# 5. 채우기 및 외곽선 그리기
# extend='max'는 가장 높은 레벨 이상의 영역도 마지막 색으로 채우라는 의미
ax.contourf(X, Y, Z, levels=levels, colors=custom_colors, extend='max')
# 굵은 검은색 외곽선 추가
ax.contour(X, Y, Z, levels=levels, colors='black', linewidths=2)

# 6. 마무리 (축 제거 등)
ax.axis('off')
ax.set_aspect('equal')
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

# 7. SVG 저장
plt.savefig('multimode_pattern.svg', format='svg', bbox_inches='tight', pad_inches=0)

# 결과 출력
plt.show()
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
import colorsys
import matplotlib.colors as mc

# -----------------------------------------------------------
# [Helper] 색상 밝기 조절 함수
# -----------------------------------------------------------
def adjust_lightness(color, amount=0.5):
    """
    입력된 색상의 밝기(Lightness)를 조절합니다.
    amount < 1 : 어둡게 (0.5면 50% 밝기)
    amount > 1 : 밝게
    """
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = mc.to_rgb(c)
    c = colorsys.rgb_to_hls(*c)
    return colorsys.hls_to_rgb(c[0], max(0, min(1, amount * c[1])), c[2])

# -----------------------------------------------------------
# 시각화 함수
# -----------------------------------------------------------
def plot_horizontal_compute_metrics(df, model_order=None):
    # --- 스타일 설정 ---
    plt.style.use('seaborn-v0_8-whitegrid')
    plt.rcParams.update({
        'font.family': 'monospace', 
        'font.size': 14,       
        'axes.labelsize': 16, 
        'xtick.labelsize': 20, 
        'ytick.labelsize': 20, 
        'figure.titlesize': 40,
        'axes.titlesize': 18
    })

    # --- 색상 매핑 ---
    full_palette = sns.color_palette("husl", 12)
    color_index_map = {
        'Flow-BC': 8, 'MF-BC': 7, 
        'LPS': 0, 'DSRL': 1, 'dataset': 'gray'
    }

    if model_order is not None:
        hue_order = model_order
    else:
        hue_order = sorted(df['Method'].unique())
    
    custom_palette = {}
    for m in hue_order:
        idx = color_index_map.get(m, 5) 
        custom_palette[m] = full_palette[idx % 12]

    # --- 캔버스 설정 ---
    fig, axes = plt.subplots(1, 3, figsize=(20, 6)) # 너비를 조금 더 넉넉하게
    
    # 단위 변환 (MB -> GB, ms -> s 등은 필요에 따라 조정)
    # df 복사본을 만들어 원본 데이터 보존
    df_plot = df.copy()
    df_plot['perf/vram_usage_mb'] = df_plot['perf/vram_usage_mb'] / 1000 # MB -> GB 가정
    df_plot['perf/train_step_time_sec'] = df_plot['perf/train_step_time_sec'] / 1000 # MB -> GB 가정

    metrics_info = [
        ("perf/vram_usage_mb", "VRAM Usage (GB)", "VRAM Consumption (GB ↓)"),
        ("perf/train_step_time_sec", "Train (s)", "Training Latency (sec ↓)"),
        ("perf/inference_latency_ms", "Inference (ms)", "Inference Latency (ms ↓)")
    ]

    for i, (col, label, title) in enumerate(metrics_info):
        ax = axes[i]
        
        # 1. 막대 그래프 그리기 (기본 edgecolor는 무시)
        sns.barplot(
            data=df_plot,
            x=col, y="Method", 
            palette=custom_palette,
            order=hue_order,
            errorbar='sd',
            capsize=0.1,      
            linewidth=0, # 초기 테두리는 없음 (아래에서 직접 설정)
            ax=ax, alpha=1.0, 
            width=0.6,
            orient='h'
        )

        # 2. X축 범위 설정
        max_val = df_plot[col].mean() + df_plot[col].std()
        x_limit = max_val * 1.5 
        ax.set_xlim(0, x_limit)

        # 3. [핵심] 테두리 색상 변경 및 텍스트 추가
        for container in ax.containers:
            for bar in container:
                if isinstance(bar, mpatches.Rectangle):
                    # (A) 현재 막대 색상 가져오기
                    face_color = bar.get_facecolor()
                    
                    # (B) 색상을 더 어둡게 변환 (amount=0.6 -> 60% 밝기)
                    edge_color = adjust_lightness(face_color, amount=0.6)
                    
                    # (C) 테두리 적용
                    bar.set_edgecolor(edge_color)
                    bar.set_linewidth(5.0) # 테두리 두께

                    # (D) 수치 텍스트 표시
                    bar_w = bar.get_width()
                    bar_y = bar.get_y() + bar.get_height() / 2
                    
                    if bar_w > 0:
                        ax.text(
                            bar_w + (x_limit * 0.05), # 간격 5%로 유지
                            bar_y, 
                            f'{bar_w:.2f}',
                            va='center', ha='left', 
                            fontsize=30, fontweight='bold',
                            fontfamily='monospace'
                        )

        # 4. 스타일 정리
        ax.set_title(title, fontweight='bold', pad=15, fontsize=30)
        ax.set_xlabel("") # X축 라벨 제거 (깔끔하게)
        ax.set_ylabel("") 
        
        if i > 0:
            ax.set_yticklabels([])
        
        ax.grid(axis='x', linestyle='-', alpha=0.5, linewidth=1, zorder=0.5)
        ax.set_axisbelow(True)

    plt.tight_layout()
    
    filename = "style_plot_compute_dark_border.pdf"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved {filename}")
    plt.show()

# -----------------------------------------------------------
# 실행부
# -----------------------------------------------------------
# 데이터 준비 (가정)
df_compute = vram_df[vram_df['_step'] >= 100].copy()
name_map = {'mflql': 'LPS', 'meanflow': 'MF-BC', 'dsrl': 'DSRL', 'flow': 'Flow-BC'}
df_compute['Method'] = df_compute['group_id'].map(lambda x: name_map.get(x, x))

# 순서 정의
my_model_order = ['Flow-BC', 'MF-BC', 'DSRL', 'LPS'] 

# 실행
plot_horizontal_compute_metrics(df_compute, model_order=my_model_order)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ---------------------------------------------------------
# 1. 데이터 준비 (CSV 파일이 있다면 pd.read_csv 사용)
# ---------------------------------------------------------
# 예시를 위해 데이터를 직접 생성합니다. 실제 사용 시에는 pd.read_csv('your_log.csv')를 쓰세요.
df = online_df
# [중요] 사용자의 데이터 수정 요청 반영
# DSRL의 마지막 스텝(약 5000) 근처의 성공률을 1 -> 0으로 강제 수정
# 실제 데이터 로드 후 이 코드를 실행하세요.
last_dsrl_idx = df[(df['group_id'] == 'dsrl')].index[-1]
df.loc[last_dsrl_idx, 'eval/success'] = 0.0

# ---------------------------------------------------------
# 2. 데이터 전처리 (Moving Average)
# ---------------------------------------------------------
WINDOW_SIZE = 20

# 그룹별로 정렬 후 Rolling Mean 적용
df = df.sort_values(by=['group_id', '_step'])
df['smoothed_success'] = df.groupby('group_id')['eval/success'].transform(
    lambda x: x.rolling(window=WINDOW_SIZE, min_periods=1).mean()
)

# ---------------------------------------------------------
# 3. 플롯 스타일 설정 (논문 테마)
# ---------------------------------------------------------
# 논문에 사용된 색상 코드를 여기에 입력하세요. (임시값 지정됨)
colors = {
    'lql': '#F77189',  # LPS (Ours) - 보통 붉은/주황 계열
    'dsrl': '#E68332', # DSRL - 보통 파란 계열
    'offline_lps': '#F77189',
    'offline_dsrl': '#E68332'
}

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    'font.family': 'monospace', 
    'font.size': 18,       
    'axes.labelsize': 20, 
    'xtick.labelsize': 18, 
    'ytick.labelsize': 16, 
    'figure.titlesize': 24,
    'legend.fontsize': 16
})

# ---------------------------------------------------------
# 4. 그리기 (Subplot 중 하나라고 가정)
# ---------------------------------------------------------
fig, ax = plt.subplots(figsize=(7.5, 5)) # 사이즈는 조정 가능

# (1) Offline Baselines (점선)
ax.axhline(y=0.30, color=colors['offline_lps'], linestyle='--', alpha=0.6, linewidth=1.5, label='LPS (Offline)')
ax.axhline(y=0.25, color=colors['offline_dsrl'], linestyle='--', alpha=0.6, linewidth=1.5, label='DSRL (Offline)')

# (2) Online Adaptation Curves (실선)
# LPS (Ours)
sns.lineplot(
    data=df[df['group_id'] == 'lql'],
    x='_step',
    y='smoothed_success',
    color=colors['lql'],
    label='LPS (Online)',
    ax=ax,
    linewidth=3
)

# DSRL
sns.lineplot(
    data=df[df['group_id'] == 'dsrl'],
    x='_step',
    y='smoothed_success',
    color=colors['dsrl'],
    label='DSRL (Online)',
    ax=ax,
    linewidth=3
)

# ---------------------------------------------------------
# 5. 세부 디자인 (축, 라벨, 범례)
# ---------------------------------------------------------
ax.set_xlabel('Environment Steps')
ax.set_ylabel('Success Rate')
ax.set_title('Online Fine-tuning Performance', fontweight='bold',)

# Y축 범위 (0 ~ 1.0 혹은 여유 있게 1.1)
ax.set_ylim(-0.05, 0.8)
ax.set_xlim(0, 5000)

# 범례 정리 (순서 및 위치)
# 중복 제거 및 커스텀 오더링이 필요할 수 있음
handles, labels = ax.get_legend_handles_labels()
# 원하는 순서: LPS Online -> DSRL Online -> LPS Offline -> DSRL Offline
order = [2, 3, 0, 1] 
ax.legend([handles[idx] for idx in order], [labels[idx] for idx in order], 
          loc='upper left', frameon=True, framealpha=0.9)

# 불필요한 테두리 제거 (Top, Right)
sns.despine()

plt.tight_layout()
# plt.show()
plt.savefig('online_finetuning.svg', bbox_inches='tight') # 저장 시
```
